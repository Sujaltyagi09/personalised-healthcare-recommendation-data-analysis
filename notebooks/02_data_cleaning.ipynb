```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("healthcare_dataset.csv")
```

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())
```

```python
df_clean['Test Results'] = df_clean['Test Results'].fillna("Not Available")
```

```python
df_clean = df_clean[df_clean['Age'] > 0]
```

```python
df_clean = df_clean[df_clean['Billing Amount'] > 0]
```

```python
df_clean['Date of Admission'] = pd.to_datetime(df_clean['Date of Admission'], errors='coerce')
df_clean['Discharge Date'] = pd.to_datetime(df_clean['Discharge Date'], errors='coerce')
```

```python
df_clean[['Date of Admission', 'Discharge Date']].info()
```

```python
text_cols = ['Gender', 'Medical Condition', 'Admission Type', 'Blood Type']

for col in text_cols:
    df_clean[col] = df_clean[col].astype(str).str.strip().str.title()
```

```python
df_clean.isna().sum()
```

```python
df_clean.describe()
```

```python
df_clean.info()
```

```python
df_clean.to_csv("healthcare_cleaned.csv", index=False)
```

### Data Cleaning Summary
- Removed duplicate records
- Handled missing values in Age and Test Results
- Removed invalid Age and Billing Amount records
- Converted date columns to datetime format
- Standardized categorical text fields
