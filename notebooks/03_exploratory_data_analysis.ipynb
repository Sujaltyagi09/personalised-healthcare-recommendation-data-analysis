```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("healthcare_cleaned.csv")
```

```python
df.shape

df.head()

df.info()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.title("Age Distribution of Patients")
plt.tight_layout()
plt.savefig("age_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Number of Patients")
plt.title("Gender Distribution")
plt.tight_layout()
plt.savefig("gender_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Medical Condition'].value_counts().plot(kind='bar')
plt.xlabel("Medical Condition")
plt.ylabel("Number of Patients")
plt.title("Distribution of Medical Conditions")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Medical Condition', y='Age', data=df)
plt.xticks(rotation=45)
plt.title("Age Distribution Across Medical Conditions")
plt.tight_layout()
plt.savefig("disease_vs_age.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['Admission Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Admission Type Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Billing Amount'], bins=30)
plt.xlabel("Billing Amount")
plt.ylabel("Number of Patients")
plt.title("Billing Amount Distribution")
plt.tight_layout()
plt.savefig("billing_analysis.png")
plt.show()
```

```python
avg_bill = df.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending=False)
avg_bill.plot(kind='bar')
plt.xlabel("Medical Condition")
plt.ylabel("Average Billing Amount")
plt.title("Average Billing Amount by Medical Condition")
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['Medical Condition'], df['Gender'])
```

### Key EDA Insights
- Majority of patients belong to the adult and senior age groups
- Certain medical conditions show higher average billing amounts
- Emergency admissions contribute significantly to high-cost cases
- Some diseases are more prevalent in specific age groups
- Billing amount distribution is right-skewed, indicating cost outliers
